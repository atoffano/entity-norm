```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set(style='ticks', context='talk')
data = pd.DataFrame({'accuracy' : [0.902, 0.899, 0.914, 0.901, 0.905, 0.911, 0.901, 0.904, 0.905, 0.904], 'type' : ''})
sns.swarmplot(x='type', y='accuracy', data=data)
sns.despine()
```
