```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ['Author', 'Reproduction']
y = [91.1, 90.4687500000]
y_errormin =[0, abs(y[1] - 90.1386356366)]
y_errormax =[0, abs(y[1] - 90.7988643634)]
y_error =[y_errormin, y_errormax]
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']
plt.ylim(90,92)
plt.bar(x, y, yerr = y_error, color=[colors[0], colors[2]])
plt.show() 
```
